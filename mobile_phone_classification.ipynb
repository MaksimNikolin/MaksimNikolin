```python
mobile_train = pd.read_csv("/kaggle/input/mobile-price-classification/train.csv")
mobile_test = pd.read_csv("/kaggle/input/mobile-price-classification/test.csv")
```

```python
X_train = mobile_train.drop(['clock_speed', 'm_dep', 'n_cores', 'price_range'], axis = 1)
y_train = mobile_train.price_range
X_test = mobile_test.drop(['id', 'clock_speed', 'm_dep', 'n_cores'], axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
np.random.seed(0)
parametr = {'n_estimators' : range (2, 10), 'max_depth' : range (5, 10), 'min_samples_leaf' : range (1, 5), 'min_samples_split' : range (2, 5)}
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
gridsearchCV_clf = GridSearchCV(rf, parametr, cv=5)
gridsearchCV_clf.fit(X_train, y_train)
```

```python
gridsearchCV_clf.best_params_
```

```python
best_clf = gridsearchCV_clf.best_estimator_
best_clf.score(X_train, y_train)
```

```python
predict = rf.predict(X_test)
```
